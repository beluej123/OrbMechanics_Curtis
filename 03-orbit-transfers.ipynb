## Hohmann transfer<font size=2>

The Hohmann transfer orbit is a solution to transfer between two circular orbits, via an elliptic transfer orbit, using the least possible velocity change ($\Delta v$). It was first demonstrated by Hohmann in 1925, before the rocket technology to orbit existed.

One Hohmann transfer example begins with a lower-altitude circular "parking orbit", then designs an elliptical transfer orbit to reach a higher-altitude circular orbit. This involves two impulsive velocity burns ($\Delta v$). The initial burn takes place at the transfer orbit periapsis. The second burn takes place at the transfer orbit aposis - the apoapsis kick.

**The problem:** Given the original orbit's periapsis and apoapsis distances ($r_{p1}$ and $r_{a1}$),
and the apoapsis distance of the desired transfer orbit (which also may be the radius of the final circular orbit) $r_{a2}$, find the required instantaneous change in velocity $\Delta v$ to enter the elliptical transfer orbit.

The change in velocity needed is the difference in periapsis velocities of the two orbits.  
For orbit 1:

$$
\begin{align}
e_1 &= \frac{r_{a1} - r_{p1}}{r_{a1} + r_{p1}} \\
r_{p1} &= \frac{h_1^2 / \mu}{1 + e_1} \\
\rightarrow h_1 &= \sqrt{2 \mu} \sqrt{\frac{r_{p1} r_{a1}}{r_{p1} + r_{a1}}} \\
v_{p1} &= \frac{h_1}{r_{p1}} \;,
\end{align}
$$

where $v_{p1}$ is the velocity of orbit 1 at its periapsis.

For orbit 2:

$$
\begin{align}
h_2 &= \sqrt{2 \mu} \sqrt{\frac{r_{p2} r_{a2}}{r_{p1} + r_{a2}}} \\
v_{p2} &= \frac{h_2}{r_{p2}} \;,
\end{align}
$$

where $r_{p2} = r_{p1}$ is the periapsis distance, shared between the orbits.
Then, the instantaneous velocity change needed is

$$
\Delta v = v_{p2} - v_{p1} \;.
$$

Orbit 2 is an elliptical transfer orbit, used to reach a higher circular orbit via an "apoapsis kick", which is a second velocity change at the apoapsis of the elliptical orbit. We can determine the velocity change needed to enter a circular orbit with a radius matching $r_{a2}$:

$$
\begin{align}
h_3 &= \sqrt{2 \mu} \sqrt{\frac{r_{a2} r_{a2}}{r_{a2} + r_{a2}}} = \sqrt{\mu r_{a2}} \\
v_{3} &= \frac{h_3}{r_{a2}} \;, \\
\Delta v &= v_3 - v_{a2}
\end{align}
$$


## Apse line rotation<font size=2>

A more general orbital transfer is an apse line rotation, which involves an instantaneous velocity change applied at some location in one orbit and leads to a new orbit with a different eccentricity and periapsis/apoapsis distances.

**Problem:** Given initial perigee distance $r_{p1}$, initial apogee distance $r_{a1}$,
desired perigee distance $r_{p2}$, desired apogee distance $r_{a2}$,
and apse rotation angle $\eta$.

Find: true anomaly of the burn $\theta_1$, required instantaneous velocity change $\Delta v$,
and the required direction/angle of thrust from the local horizontal direction.

Solution steps:

1. Calculate $e_1$, $e_2$, $h_1$, $h_2$:

   $$
   \begin{align}
   e_1 &= \frac{r_{a1} - r_{p1}}{r_{a1} + r_{p1}} \\
   r_p &= \frac{h_1^2 / \mu}{1 + e_1} \\
   \rightarrow h_1 &= \sqrt{ r_p (1 + e_1) \mu} \;,
   \end{align}
   $$

   then repeat for $e_2$ and $h_2$.

2. Solve for the true anomalies in each orbit of the intercept, $\theta_1$ and $\theta_2$:

$$
\begin{align}
\eta &= \theta_1 - \theta_2 \rightarrow \theta_2 = \theta_1 - \eta \\
r_{1I} &= \frac{h_1^2 / \mu}{1 + e_1 \cos \theta_1} = r_{2I} = \frac{h_2^2 / \mu}{1 + e_2 \cos \theta_2}
\end{align}
$$

Following some math, we can solve for the angles:

$$
\begin{align}
\theta_1 &= \phi + \cos^{-1} \left( \frac{c}{a} \cos \phi \right) \\
\theta_2 &= \theta_1 - \eta \\
\text{where} \quad \phi &= \tan^{-1} \left( \frac{b}{a} \right) \\
a &= e_1 h_2^2 - e_2 h_1^2 \cos \eta \\
b &= -e_2 h_1^2 \sin \eta \\
c &= h_1^2 - h_2^2 \;,
\end{align}
$$

where $\phi$ gives the direction of the local horizontal.

3. Calculate the position and velocity in the initial orbit:

$$
\begin{align}
v_{theta 1} &= \frac{h_1}{r_1} \\
v_{r 1} &= \frac{\mu}{h_1} e_1 \sin \theta_1 \\
v_1 &= \sqrt{ v_{r1}^2 + v_{\theta 1}^2 } \\
\gamma_1 &= \tan^{-1} \left( \frac{v_{r1}}{v_{\theta 1}} \right)
\end{align}
$$

4. Calculate the position and velocity in the desired orbit:

$$
\begin{align}
v_{theta 1} &= \frac{h_2}{r_2} \\
v_{r 2} &= \frac{\mu}{h_2} e_2 \sin \theta_2 = \frac{\mu}{h_2} e_2 \sin (\theta_1 - \eta) \\
v_2 &= \sqrt{ v_{r2}^2 + v_{\theta 2}^2 } \\
\gamma_2 &= \tan^{-1} \left( \frac{v_{r2}}{v_{\theta 2}} \right)
\end{align}
$$

5. Calculate the instantaneous change in velocity needed:

$$
\Delta v = \sqrt{ v_1^2 + v_2^2 - 2 v_1 v_2 \cos \left( \theta_2 - \theta_1 \right)} \;,
$$

at the intercept location.

6. Calculate the direction/angle of thrust:

$$
I = \tan^{-1} \left( \frac{v_{r2} - v_{r1}}{v_{\theta 2} - v_{\theta 1}} \right) + 180^{\circ}
$$


## Cost of transfers<font size=2>

The cost of an orbit transfer, in terms of the mass of propellant needed, depends on a few factors:

- propellant used
- thrust developed
- rate of fuel consumption ($\dot{m}$)

The **specific impulse**, given in units of seconds, is a performance parameter that quantifies the amount of thrust developed to the mass flow rate of propellant required, where higher values indicate (theoretically) better performance:

$$
I_{sp} = \frac{T}{\dot{m} g_0} \;,
$$

where $T$ is the thrust force (N), $\dot{m}$ is the mass flow rate (kg/s),
and $g_0$ is the acceleration due to gravity on the ground (9.81 m/s$^2$).
The mass flow rate is directly related to the time rate of change of vehicle mass:
$\dot{m} = -\frac{dm}{dt}$.

Based on a free-body diagram, we can write the equation of motion in the tangential (axial) direction:

$$
m a_t = T - D - m g \sin \gamma = m \frac{dV}{dt} \;,
$$

where in space we can neglect the effects of drag, or $D \approx 0$, due to the extremely rarified
atmosphere (for highly accurate calculations, we _will_ need to account for the perturbuations
due to atmospheric drag); we can also neglect the weight due to reduced gravity, so $m g \approx 0$.
We can also assume thrust is constant. Then:

$$
\begin{align}
T = m \frac{dV}{dt} &= I_{sp} g_0 \left( - \frac{dm}{dt} \right) \\
- I_{sp} g_0 \frac{dm}{dt} &= m \frac{dV}{dt} \\
\int_{m_0}^{m_f} - I_{sp} g_0 \frac{dm}{m} &= \int_0^{\Delta V} dV \\
\Delta v &= - I_{sp} g_0 \ln \left( \frac{m_f}{m_0} \right) \\
\rightarrow \Delta m &= m_0 \left( 1 - e^{-\Delta V / \left( I_{sp} g_0 \right)} \right) \;,
\end{align}
$$

where $\Delta m = m_0 - m_f$, which is the **ideal rocket equation**,
originally developed by Tsiolkovsky.
Given the specific impulse and initial mass of a rocket, this tells us the
change in mass needed (via exhaust of propellant) to achieve a particular change
in velocity.


## Finite burn times<font size=2>

While some burns can reasonably be approximated as instantaneous, other orbit transfers
will involve or require finite burn times.
These involve a burn of the engine starting at some time (and some true anomaly in the initial orbit), and leads to a different orbit, similar to an apse line rotation.

To determine how a finite burn time affects an object's orbit, we need to solve the orbital equation of motion, but with a thrust force:

$$
\vec{\ddot{r}} + \frac{\mu}{r^3} \vec{r} = \frac{\vec{F}_T}{m} = \frac{T \, \vec{v}}{m v} \;,
$$

assuming that the thrust force $\vec{F}_T$, with magnitude $T$, is in the same direction
as the velocity $\vec{v}$,
along with an equation for the changing mass of the vehicle, as it expels propellant:

$$
\frac{dm}{dt} = -\frac{T}{I_{sp} g_0} \;.
$$

This can be solved by decomposing into a system of first-order equations, and integrating numerically:

$$
\begin{align}
z_1 &= x \quad \dot{z}_1 = z_4 \\
z_2 &= y \quad \dot{z}_2 = z_5 \\
z_3 &= z \quad \dot{z}_3 = z_6 \\
z_4 &= \dot{x} \quad \dot{z}_4 = \frac{-\mu}{r^3} z_1 + \frac{T z_4}{z_7 v} \\
z_5 &= \dot{y} \quad \dot{z}_5 = \frac{-\mu}{r^3} z_2 + \frac{T z_5}{z_7 v} \\
z_6 &= \dot{z} \quad \dot{z}_6 = \frac{-\mu}{r^3} z_3 + \frac{T z_6}{z_7 v} \\
z_7 &= m \quad \dot{z}_7 = -\frac{T}{I_{sp} g_0} \;,
\end{align}
$$

where

$$
\begin{align}
r &= \sqrt{z_1^2 + z_2^2 + z_3^2} \\
v &= \sqrt{z_4^2 + z_5^2 + z_6^2} \;.
\end{align}
$$

To solve a problem where the burn time $\Delta t$ to reach a particular orbit is unknown,
an iterative approach can be used.

The conditions going from planetocentric orbit to the heliocentric orbit or vice-versa are called the “patch conditions.” \
The key patch condition is to relate the velocity of the vehicle with respect to the planet to the velocity of the vehicle with respect to the Sun. \
The difference of the two is the velocity of the planet with respect to the Sun. Since velocities are vectors, we need to account for magnitudes and directions.

### Heliocentric Phase


```python
# p.290+
import math

mu_sun = 1.327e11  # [km^3/s^2]
mu_earth = 398600  # [km^3/s^2]
r_earth_orb = 149.597893e6  # earth orbit [km]
r_mars_orb = 227.9e6  # mars orbit [km]

# Earth heliocentric velocity
v_e = math.sqrt(mu_sun / r_earth_orb)
print(f"earth orbit velocity, v_e = {v_e:.6g} [km/s]")

a_t = (r_earth_orb + r_mars_orb) / 2
v_pt = math.sqrt(mu_sun * ((2 / r_earth_orb) - (1 / a_t)))
print("values with respect to system bary-center")
print(f"semi-major axis, transfer, a_t = {a_t:.6g} [km]")
print(f"velocity @ periapsis transfer, v_pt = {v_pt:.6g} [km/s]")

# transfer orbit @apoapsis at MArs
v_at = math.sqrt(mu_sun * ((2 / r_mars_orb) - (1 / a_t)))
print(f"velocity @ apoapsis transfer, v_at = {v_at:.6g} [km/s]")
v_inf = v_pt - v_e
print(f"velocity @ Earth SOI, v_inf = {v_inf:.6g} [km/s]")
```
